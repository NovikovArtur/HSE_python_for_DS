8. Создайте новый датафрейм, в который войдут 10 самых возрастных клиентов, которые работают. Оставьте в новом датафрейме (помимо индекса) только колонки age, marital education, housing, loan

10. Какие два столбца исходного датафрейма наиболее скоррелированные?

```python
df_corr = df.corr()                                                       #создаем новый дф, заполненный коррелируемыми значениями
np.fill_diagonal(df_corr.values, np.nan)                                  #удаляем значения по диагонали, так как они всегда равны 1 и будут нам мешать

maximum = df_corr.max().max()                                             #находим значение максимальной корреляции

column = df_corr.eq(maximum).any()                                        #методом eq заполняем дф булевыми значениями, которые отражают, есть ли в данном столбце нужный нам элемент, затем методом any(), мы находим все значения True
itog_column = column[column].index[0]                                     #берем 1-й столбец, где мы находим true

row = df_corr.eq(maximum).any(axis=1)                                     #аналогично столбцам, только для строк
itog_row = row[row].index[1]

print("Столбец с максимальным значением корреляции: ", itog_column)
print("Строка с максимальным значением корреляции: ", itog_row)
```
